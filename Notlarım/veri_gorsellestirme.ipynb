Bar Plot (Sütun Grafiği) -> Kategorik değişkenler

```python
sns.barplot(df["cut"])
```

```python
sns.barplot(x= "cut", y = "price", data= df)
```

```python
sns.barplot(x= "cut", y= "price", hue= "color", data= df)
```

Histogram Ve Yoğunluk Grafikleri -> Sürekli değişkenler

```python
sns.displot(df["price"])
```

```python
sns.displot(data= df["price"], kind= "kde")
```

```python
sns.kdeplot(df["price"])
```

Box Plot (Kutu Grafik) -> Sürekli Değişkenler

```python
sns.boxplot(df2["total_bill"])
```

```python
sns.boxplot(x= "time", y= "total_bill", data= df2)
```

```python
sns.boxplot(x= "day", y= "total_bill", hue= "sex", data= df2)
```

Violin Grafiği -> Sürekli Değişkenler

```python
sns.catplot(kind= "violin", data= df2["total_bill"])
```

```python
sns.catplot(x= "day", y= "total_bill", kind= "violin", data= df2)
```

```python
sns.catplot(x= "day", y= "total_bill", hue= "sex", kind= "violin", data= df2)
```

Korelasyon Grafiği -> Sürekli değişkenler arasındaki ilişki

```python
sns.scatterplot(x= "total_bill", y= "tip", data= df2)
```

```python
sns.scatterplot(x= "total_bill", y= "tip", hue= "time", data= df2)
```

```python
sns.scatterplot(x= "total_bill", y= "tip", hue= "day", data= df2)
```

```python
sns.lmplot(x= "total_bill", y= "tip", data= df2)
```

```python
sns.lmplot(x= "total_bill", y= "tip", hue= "time", data= df2)
```

```python
sns.lmplot(x= "total_bill", y= "tip", hue= "day", data= df2)
```
